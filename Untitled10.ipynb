# Business Understanding

As shown by the ongoing Coronavirus pandemic, the need for public acceptance of emergency health direction--such as getting vaccinated--is often met with an obstinant and suspicious populace. 

If we had a means of predicting who or what kind of person is likely to resist vaccination, we might be able to more effectively target our public relations and outreach campaigns.

Here we develop a predictive model to help elucidate who is likely to accept a novel vaccination and present our results to a government agency concerned with promoting vaccination during pandemic events.

The following two functions plot a model's ROC curve. The first works with models that use decision functions, the second for those that do not.

```python
def plot_roc(model, X_train, X_test, y_train, y_test):

    y_train_score = model.decision_function(X_train)
    y_test_score = model.decision_function(X_test)
    
    train_fpr, train_tpr, train_thresholds = roc_curve(y_train, y_train_score)
    test_fpr, test_tpr, test_thresholds = roc_curve(y_test, y_test_score)
    
    print('Train AUC: {}'.format(auc(train_fpr, train_tpr)))
    print('Test AUC: {}'.format(auc(test_fpr, test_tpr)))
    print('------------------------------------')
    
    plt.figure(figsize=(10, 8))
    lw = 2
    plt.plot(train_fpr, train_tpr, color = 'blue',
             lw=lw, label='Train ROC curve')
    plt.plot(test_fpr, test_tpr, color = 'darkorange',
             lw=lw, label='Test ROC curve')
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([-.01, 1.0])
    plt.ylim([0.0, 1.05])
    plt.yticks([i/20.0 for i in range(21)], fontsize = 10)
    plt.xticks([i/20.0 for i in range(21)], fontsize = 10)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()
```

```python
def plot_roc_2(model, X_train, X_test, y_train, y_test):

    y_train_score = model.predict_proba(X_train)
    y_test_score = model.predict_proba(X_test)
    
    train_fpr, train_tpr, train_thresholds = roc_curve(y_train, y_train_score [:,1])
    test_fpr, test_tpr, test_thresholds = roc_curve(y_test, y_test_score [:,1])
    
    print('Train AUC: {}'.format(auc(train_fpr, train_tpr)))
    print('Test AUC: {}'.format(auc(test_fpr, test_tpr)))
    print('------------------------------------')
    
    plt.figure(figsize=(10, 8))
    lw = 2
    plt.plot(train_fpr, train_tpr, color = 'blue',
             lw=lw, label='Train ROC curve')
    plt.plot(test_fpr, test_tpr, color = 'darkorange',
             lw=lw, label='Test ROC curve')
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([-.01, 1.0])
    plt.ylim([0.0, 1.05])
    plt.yticks([i/20.0 for i in range(21)], fontsize = 10)
    plt.xticks([i/20.0 for i in range(21)], fontsize = 10)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()
```

## Baseline Model: Logistic Regression

Our models will be used in a pipeline that first scales the data with standardscaler, then fits the scaled train test split data to the specified algorithm. We will fit on both training and evaluation sets. Logistic regression serves as a great baseline for binary classification problems because it yields interpretable and quick outputs. 

It's a great and simple way to begin to look at feature importances and potential ways to tune parameter and hyperparameters in subsequent iterations.

```python
scaler = StandardScaler()
model = LogisticRegression(random_state=5)
pipeline_lr = Pipeline(steps=[('ss', scaler),
                              ('m', model)])
y_pred = pipeline_lr.fit(X_train, y_train).predict(X_test)
print('Cross Validation ROC-AUC Score: ',cross_val_score(pipeline_lr, X_train, y_train, scoring='roc_auc',cv=5).mean())
plot_confusion_matrix(pipeline_lr, X_test, y_test)
plot_roc(pipeline_lr, X_train, X_test, y_train, y_test)
```

Our first vanilla model using logistic regression performed well--very well for the majority class--but the model had a bit of trouble correctly identifying the minority cases. 

The evaluation metric AUC (area under the curve) represents a model's discriminatory performance. An AUC of would indicate 100% accurate predictive ability--that is, a model that classifies the data correctly 100% of the time. An AUC of 0.5 indicates a model no better at prediction than random chance.

Our baseline, default classifier has an AUC of 0.81. This is an excellent first result, though as is seen in the confusion matrix, our false negatives are a bit on the high side.

For our business case, this may not be an entirely horrible situation. When looking at what we're trying to predict and why, we're actually much more interested in identifying who isn't getting the vaccine, so that we can determine ways to increase vaccine intake. And given the numbers, our model is quite good at predicting who won't get the jab.

Of course, logistic regression is just one of many algorithms we can try. Next we will run through several different, more complicated algorithms and see where their baselines take us.

## Baseline Model: Support Vector Machine

Support vector machines are classifiers that attempt to differentiate classes by maximizing the width of the margin between the most ambiguous datapoints (support vectors). SVMs are great for smaller datasets and where the data are easily linearily separable. They tend to underperform with larger datasets, and as something of a geometrical construct, lack probabilistic support for their classifications.

Here we once again scale our data before fitting to a default SVM, then print our classification report, generate a confusion matrix, and plot a roc curve.

```python
scaler = StandardScaler()
model = SVC(random_state=5)
pipeline_svc = Pipeline(steps=[('ss', scaler),
                              ('m', model)])
y_pred = pipeline_svc.fit(X_train, y_train).predict(X_test)
print('Cross Validation ROC-AUC Score: ',cross_val_score(pipeline_svc, X_train, y_train, scoring='roc_auc',cv=5).mean())
plot_confusion_matrix(pipeline_svc, X_train, y_train)
plot_roc(pipeline_svc, X_train, X_test, y_train, y_test)
```

Our SVM performed well enough--almost as good as logistic regression--but has a slightly worse roc-auc score. We will likely not be pursuing a tuned SVM.

```python
from sklearn.naive_bayes import GaussianNB

scaler = StandardScaler()
model = GaussianNB()

pipeline_nb = Pipeline(steps=[('ss', scaler),
                              ('m', model)])

y_pred = pipeline_nb.fit(X_train, y_train).predict(X_test)
print('Cross Validation ROC-AUC Score: ',cross_val_score(pipeline_nb, X_train, y_train, scoring='roc_auc',cv=5).mean())
plot_confusion_matrix(pipeline_nb, X_test, y_test)
plot_roc_2(pipeline_nb, X_train, X_test, y_train, y_test)
```

The Naive Bayes algorithm does not perform quite as well as the previous two, though it seems to handle the false negatives better, of course at the cost of higher false positives. We might want to explore this one a bit more.

## Baseline Model: Decision Trees

Another type of classifying algorithm is based on the decision tree. A decision tree is an algorithm that splits a dataset into subsets based on a series of conditionals. Think "Does X has y condition?-->Y to left subset, N to right subset.

Once the data can no longer be split further, we can make our predictions.

Our first such algorithm is a single decision tree, followed by an ensemble classifier version, the random forest. The random forest being a collection of decision trees from which we make our predictions.

```python
scaler = StandardScaler()
model = DecisionTreeClassifier(random_state=5)
pipeline_dt = Pipeline(steps=[('ss', scaler),
                              ('m', model)])
y_pred = pipeline_dt.fit(X_train, y_train).predict(X_test)
print('Cross Validation ROC-AUC Score: ',cross_val_score(pipeline_dt, X_train, y_train, scoring='roc_auc',cv=5).mean())
plot_confusion_matrix(pipeline_dt, X_test, y_test)
plot_roc_2(pipeline_dt, X_train, X_test, y_train, y_test)
```

One of the big problems with tree-based algorithms is their tendency to overfit, which is plainly obvious here.

Below is our random forest.

```python
scaler = StandardScaler()
model = RandomForestClassifier(random_state=5)
pipeline_rf = Pipeline(steps=[('ss', scaler),
                              ('m', model)])
y_pred = pipeline_rf.fit(X_train, y_train).predict(X_test)
print('Cross Validation ROC-AUC Score: ',cross_val_score(pipeline_rf, X_train, y_train, scoring='roc_auc',cv=5).mean())
plot_confusion_matrix(pipeline_rf, X_test, y_test)
plot_roc_2(pipeline_rf, X_train, X_test, y_train, y_test)
```

Ok...it would seem we're overfitting on the training data. This means that the model has learned the training data too well and may not accurately generalize to new data--our test set!

These are much more in line with the test scores we found in previous trials.

### Baseline Models: Decision Trees - Gradient Boosting

The last decision tree based model we'll try is the Gradient Boosting model. Gradient Boosting models create a series of 'weak' models and then add them together. Each weak learner learns from the errors of its predecessors, and though the algorithm as a whole takes a bit more computation time, the results speak for themselves.

```python
scaler = StandardScaler()
model = GradientBoostingClassifier(random_state=5)
pipeline_gb = Pipeline(steps=[('ss', scaler),
                              ('m', model)])
y_pred = pipeline_gb.fit(X_train, y_train).predict(X_test)
print('Cross Validation ROC-AUC Score: ',cross_val_score(pipeline_gb, X_train, y_train, scoring='roc_auc',cv=5).mean())
plot_confusion_matrix(pipeline_gb, X_test, y_test)
plot_roc_2(pipeline_gb, X_train, X_test, y_train, y_test)
```

# Hyperparameter Tuning GBoost, XGBoost Classifiers

Our baseline GradientBoostingClassifier and XGBoost outperformed the other algorithms we used. 

Having established our baseline, we will now use RandomizedSearchCV to optimize the algorithm's hyperparameters to end up with the highest AUC score we can. 

We will also use SMOTE to deal with the unbalanced data and see how that affects our metrics. First is our baseline model with SMOTE resampled training and testing data.

```python

smote = SMOTE(random_state=5)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

scaler = StandardScaler()
model = GradientBoostingClassifier()
pipeline = Pipeline(steps=[('ss',scaler),
                           ('m',model)])

gb = pipeline.fit(X_train_res, y_train_res)
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=5)
scores = cross_val_score(gb, X_train_res, y_train_res, scoring='roc_auc', cv=cv, n_jobs=-1)
print(scores.mean())
```

That's definitely an improvement!

For GradientBoostingClassifier we will focus on the most important hyperparameters:
1. max_depth: The number of layers of nodes/leaves in the tree.
2. learning rate: Weight assigned to each tree generated after boosting.
3. min_samples_split: The mininmum number of samples required to split an internal node.
4. min_samples_leaf: The minimum number of samples required to be in a leaf.
5. max_features: How many features will be considered while splitting--randomly selected.
6. subsample: Percentage of observations considered for each tree.
7. n_estimators: How many trees are made

Note that I have commented out several hyperparameters to cut back on computational time. I've left them in to remind myself that they may be important.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform as sp_randFloat
from scipy.stats import randint as sp_randInt

parameters = {
    "max_depth": sp_randInt(2,10),
    'learning_rate': sp_randFloat(),
    #'min_samples_split':sp_randFloat(),
    #'min_samples_leaf':sp_randFloat(),
    "max_features":["log2","sqrt"],
    "subsample": sp_randFloat(),
    #"n_estimators":sp_randInt(100,1000)
}
clf = GradientBoostingClassifier()
clf.fit(X_train_res, y_train_res)
cv = RandomizedSearchCV(clf, param_distributions=parameters, n_iter=100, cv=5, n_jobs=-1, scoring='roc_auc')
cv.fit(X_train_res, y_train_res)
print(cv.best_params_)
print(cv.best_score_)
```
